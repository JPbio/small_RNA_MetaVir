## Build Data Frame

#### Auxiliary methods

Masks

```python

# File size masks
bytes_2_kb = lambda size_bytes: size_bytes / (1024)
bytes_2_mb = lambda size_bytes: size_bytes / (1024 * 1024)
bytes_2_gb = lambda size_bytes: size_bytes / (1024 * 1024 * 1024)

bytes_2_gb_str = lambda size_bytes: "{:.2f} GB".format(bytes_2_gb(size_bytes))
bytes_2_mb_str = lambda size_bytes: "{:.2f} MB".format(bytes_2_mb(size_bytes))
bytes_2_kb_str = lambda size_bytes: "{:.2f} KB".format(bytes_2_kb(size_bytes))

# Scientific notation masks
mask_thousands = lambda x: "{:.2f}".format(x / 1000)
mask_million = lambda x: "{:.2f}".format(x / 1000 / 1000)
```

Report rows building

```python
def add_masks(df_row: pd.DataFrame) -> pd.DataFrame:

    # Input size masks
    df_row[f'{col_input} [GB]'] = df_row[col_input].apply(bytes_2_gb_str)
    df_row[f'{col_input} [MB]'] = df_row[col_input].apply(bytes_2_mb_str)
    df_row[f'{col_input} [KB]'] = df_row[col_input].apply(bytes_2_kb_str)

    # Output size masks
    df_row[f'{col_output} [GB]'] = df_row[col_output].apply(bytes_2_gb_str)
    df_row[f'{col_output} [MB]'] = df_row[col_output].apply(bytes_2_mb_str)
    df_row[f'{col_output} [KB]'] = df_row[col_output].apply(bytes_2_kb_str)

    # Big numbers scientific notation
    df_row[f'{col_reads} [×10^3]'] = df_row[col_reads].apply(mask_thousands)
    df_row[f'{col_reads} [×10^6]'] = df_row[col_reads].apply(mask_million)

    return df_row

def get_similarity_class_count(df_lib: pd.DataFrame, label: str) -> int:
    try:
        similarity_class_group = df_lib.groupby(col_label_similarity)[col_label_similarity].value_counts()
        return int(similarity_class_group[label].values)
    except KeyError:
        return 0
    except Exception as err:
        raise err

def get_row(df_lib: pd.DataFrame, summary_lib: dict) -> pd.DataFrame:

    class_match = lambda label, class_: \
        df_lib[ (df_lib[col_label_similarity] == label) & (df_lib[col_label_class] == class_)]

    return pd.DataFrame(
        {
            'ID': summary_lib.get('exec_id'),
            col_reads: summary_lib.get('n_read_total'),
            'Contigs': summary_lib.get('n_contig_total'),
            'Contigs >= 200nt': summary_lib.get('n_contig_gt200'),
            col_input: summary_lib.get('input_size'),
            col_output: summary_lib.get('output_size'),
            'Execution Time': summary_lib.get('exec_time'),
            
            'Server': summary_lib.get('server_name'),
            'Batch': summary_lib.get('batch_name'),
            
            'Viral': get_similarity_class_count(df_lib=df_lib, label='viral'),
            'Non Viral': get_similarity_class_count(df_lib=df_lib, label='nonviral'),
            'nohit': get_similarity_class_count(df_lib=df_lib, label='nohit'),
            
            'Viral / Viral': class_match('viral', 'viral').shape[0],
            'Viral / EVE': class_match('viral', 'eve').shape[0],
            'Non Viral / Viral': class_match('nonviral', 'viral').shape[0],
            'Non Viral / EVE': class_match('nonviral', 'eve').shape[0],
            'No Hit / Viral': class_match('nohit', 'viral').shape[0],
            'No Hit / EVE': class_match('nohit', 'eve').shape[0],
            
        },
        index=[0]
    )
```

#### Gather executions

```python
df_report = pd.DataFrame(
    columns = [
        'ID',
        'Reads',
        'Contigs', 'Contigs >= 200nt',
        'Input Size', 'Output Size',
        'Execution Time',
        'Server', 'Batch',
        'Viral', 'Non Viral', 'nohit',
        'Viral / Viral', 'Viral / EVE',
        'Non Viral / Viral', 'Non Viral / EVE',
        'No Hit / Viral', 'No Hit / EVE',
    ]
)
```

```python


# Filter out subdirectories from the list of items
run_dirs = []
for dir_item in os.listdir(path_root):
    path_run = os.path.join(path_root, dir_item)
    if os.path.isdir(path_run):
        run_dirs.append(path_run)

# Loop trough each execution
for path_run in run_dirs:

    # Get execution result files
    path_csv_classified = ''
    path_summary = ''

    for dir_item in os.listdir(path_run):
        path_file = os.path.join(path_run, dir_item)

        if path_file.endswith(suffix_summary):
            path_summary = path_file
        elif path_file.endswith(suffix_classification):
            path_csv_classified = path_file
    
    if not path_summary or not path_csv_classified:
        continue

    # Build current lib report row
    with open(path_summary, 'r') as file:
        summary_lib = yaml.safe_load(file)

    if not summary_lib.get('server_name'):
        summary_lib['server_name'] = server_r2d2

    df_lib = pd.read_csv(path_csv_classified).drop('Unnamed: 0', axis=1)
    df_row = get_row(df_lib=df_lib, summary_lib=summary_lib)

    # Add this lib parsing to report
    df_report = pd.concat([df_report, df_row], ignore_index=True)
```
